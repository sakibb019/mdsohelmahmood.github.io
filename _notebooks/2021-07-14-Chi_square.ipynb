# Chi-square test
> Implementation in python using pingouin

```python
import pandas as pd
from scipy import stats
import numpy as np
import pingouin as pg
```

```python
data = pg.read_dataset('chi2_independence')
data
```

```python
data['sex'].value_counts(ascending=True)
```

```python
data['target'].value_counts(ascending=True)
```

```python
a=data[(data['target']==1) & (data['sex']==0)]
a.shape[0]
```

```python
a=data[(data['target']==0) & (data['sex']==0)]
a.shape[0]
```

```python
expected, observed, stats = pg.chi2_independence(data, x='sex',
                                                 y='target')
```
